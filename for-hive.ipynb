```python
from pyhive import hive
import pandas as pd
```

```python
# Connect to HiveServer2 running in Docker container
conn = hive.Connection(host='localhost', port=10000, username='hive')
```

```python
# Create a cursor to interact with Hive
cursor = conn.cursor()
```

```python
# Example: Show databases (just to test the connection)
cursor.execute('SHOW DATABASES')
```

```python
# Fetch all results and print them
databases = cursor.fetchall()
print("Databases in Hive:", databases)
```

```python
# Example: Select data from a table (replace 'your_table' with an actual table name)
cursor.execute('SELECT * FROM news_data LIMIT 100')
```

```python
# Fetch results from the query
results = cursor.fetchall()
print("Query Results:", results)
```

```python
df = pd.DataFrame(results, columns=[col[0] for col in cursor.description])
```

```python
# Close the cursor and connection
cursor.close()
conn.close()
```
